```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
from pymodulon.io import *
from os import path

interim_dir = path.join('..','data','interim')
ica_data = load_json_model(path.join(interim_dir,'zmo_raw.json.gz'))
```

```python
iM = 'ETC_and_MEP'
ica_data.view_imodulon(iM).COG.value_counts().values
```

```python
ica_data.view_imodulon(iM)
```

```python
import matplotlib.pyplot as plt

# Data from the provided COG value counts
cog_categories = ica_data.view_imodulon(iM).COG.value_counts().index
counts = ica_data.view_imodulon(iM).COG.value_counts().values

# Function to format labels with both count and percentage
def func(pct, allvalues):
    absolute = int(pct / 100.*sum(allvalues))  # Calculate the absolute value from the percentage
    return f"{pct:.1f}% ({absolute})"

# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(counts, labels=cog_categories, autopct=lambda pct: func(pct, counts), startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribution of COG Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.savefig('../figures/pie_chart_of_COGs.pdf', transparent = True)
plt.show()
```
